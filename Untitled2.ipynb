```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import roc_auc_score
```

```python
train_fea = pd.read_csv('/content/training_set_features.csv')
train_fea
```

```python
train_lab = pd.read_csv('/content/training_set_labels.csv')
train_lab
```

```python
x = train_fea.drop(columns=['respondent_id'])
y = train_lab[['xyz_vaccine', 'seasonal_vaccine']]
```

```python
numeric_features = x.select_dtypes(include=['int64', 'float64']).columns
categorical_features = x.select_dtypes(include=['object']).columns
```

```python
categorical_features = categorical_features.tolist()
if 'age_group' not in categorical_features:
    categorical_features.append('age_group')

# Remove 'age_group' from numeric features if mistakenly included
numeric_features = numeric_features.tolist()
if 'age_group' in numeric_features:
    numeric_features.remove('age_group')
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Impute missing values with mean
    ('scaler', StandardScaler())  # Scale numerical features
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with most frequent category
    ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Encode categorical features
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
model = MultiOutputClassifier(RandomForestClassifier(random_state=42))
```

```python
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', model)])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
pipeline.fit(x_train, y_train)
```

```python
y_val_pred_proba = pipeline.predict_proba(x_test)
```

```python
roc_auc_xyz = roc_auc_score(y_test['xyz_vaccine'], y_val_pred_proba[0][:, 1])
roc_auc_seasonal = roc_auc_score(y_test['seasonal_vaccine'], y_val_pred_proba[1][:, 1])
```

```python
mean_roc_auc = (roc_auc_xyz + roc_auc_seasonal) / 2
print(f'Mean ROC AUC on validation set: {mean_roc_auc}')
```

```python
pipeline.fit(x, y)
```

```python
test_data = pd.read_csv('/content/test_set_features.csv')
```

```python
y_test_pred_proba = pipeline.predict_proba(test_data)
```

```python
submission = pd.DataFrame({
    'respondent_id': test_data['respondent_id'],
    'xyz_vaccine': y_test_pred_proba[0][:, 1],  # Predicting probabilities on the test dataset for submission
    'seasonal_vaccine': y_test_pred_proba[1][:, 1]
})
```
